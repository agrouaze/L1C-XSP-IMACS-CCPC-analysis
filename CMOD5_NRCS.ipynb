```python
import numpy as np
import matplotlib.pyplot as plt
import sys
path_tools = '/home1/datahome/ljessel/'
sys.path.append(path_tools)
from Scripts.wind_inversion_functions import cmod5n_forward


azi = np.arange(361)

incs = [25, 30, 35, 40]
spds = [5, 10, 15, 20]
_inc = 40
ticks = np.array([0,90,180,270,360])
#fig=plt.figure()
fig, ax = plt.subplots(figsize=(15,10)) 
for _spd in spds:
    spd = azi*0 + _spd
    inc = azi*0 + _inc
    
    sig = cmod5n_forward(spd, azi, inc, True)
    plt.plot(azi,10*np.log10(sig), label=str(_spd) + ' m/s',lw=3)
plt.grid()
plt.axis([0,360,-22,-7])
plt.ylabel('VV NRCS [dB] (from CMOD5)')
plt.xlabel('Azimuth Wind Direction [deg]')
ax.set_xticks(ticks)
plt.legend()
plt.title('Incidence Angle : ' + str(_inc) + ' deg' )
plt.savefig('NRCS_CMOD5.png')
```
